```python
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task="table-question-answering",model="google/tapas-base-finetuned-wtq")
```

```python
table=pd.read_csv("/content/Youtubertest.csv")
```

```python
import numpy as np

table.replace(r'\s+', np.nan, regex=True)
```

```python
table = table.astype(str)
```

```python
table
```

```python
query = "Pinkfong Baby Shark - Kids' Songs & Stories has how many Average Views?"
print(tqa(table=table,query=query)["answer"])
```
